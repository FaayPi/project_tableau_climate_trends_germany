```python
df_temp.head()
df_temp["Season"].unique()
```

```python
df_rain = pd.read_excel("/Users/feepieper/Library/CloudStorage/OneDrive-Persönlich/Ironhack/Module3/project/final_project/dataset_rain.xls")
df_rain.columns
```

```python
df_rain.drop(columns=['Jahreszeit', 'Date', 'Month'], inplace=True)
```

```python
df_temp.columns
df_temp.drop(columns=['Thueringen/Sachsen-Anhalt', 'Unnamed: 19'], inplace=True)
df_temp.rename(columns={'Baden-Wuerttemberg': "Baden-Württemberg", "Thueringen":"Thüringen"}, inplace=True)
df_temp["Bremen"]=df_temp["Hamburg"]
```

```python
df_rain.head()
df_rain.columns
df_rain.head()
```

```python
df_rain["Season"] = df_rain["Season"].replace("autum", "autumn")
df_rain["Season"].unique()
```

```python
df_temp_long = df_temp.melt(
    id_vars=['Season', 'Year'],
    value_vars=['Baden-Württemberg', 'Bayern', 'Berlin', 'Brandenburg', 'Bremen', 'Deutschland',
                'Hamburg', 'Hessen', 'Mecklenburg-Vorpommern', 'Niedersachsen', 'Nordrhein-Westfalen',
                'Rheinland-Pfalz', 'Saarland', 'Sachsen', 'Sachsen-Anhalt', 'Schleswig-Holstein', 'Thüringen'],
    var_name='State',
    value_name='Temperature'
)
df_temp_long.head()
```

```python
df_rain_long = df_rain.melt(
    id_vars=['Year','Season'],
    value_vars=['Baden-Württemberg', 'Bayern', 'Berlin', 'Brandenburg', 'Bremen', 'Deutschland',
                'Hamburg', 'Hessen', 'Mecklenburg-Vorpommern', 'Niedersachsen', 'Nordrhein-Westfalen',
                'Rheinland-Pfalz', 'Saarland', 'Sachsen', 'Sachsen-Anhalt', 'Schleswig-Holstein', 'Thüringen'],
    var_name='State',
    value_name='Precipitation'
)
df_rain_long.head()
df_rain_long[df_rain_long["Season"] == "winter"].head()
```

```python
df_merged = pd.merge(
    df_temp_long,
    df_rain_long,
    on=["Year","Season", "State"],
)

df_merged.head()
```

```python
df_rain_long.to_csv("df_rain_cleaned.csv", index=False)
```

```python
df_temp_long.to_csv("df_temp_cleaned.csv", index=False)

df_merged.to_csv("df_merged_final.csv", index=False)
```
